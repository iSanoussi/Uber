```python
import pandas as pd
import os
path = os.getcwd()
print(path)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
df = pd.read_csv("uber-raw-data-apr14.csv", delimiter=",")
```

```python
msno.bar(df)
```

```python
df.describe()
```

```python
df
```

```python
df.info()
```

```python
df["Date/Time"] = df["Date/Time"].map(pd.to_datetime)
```

```python
df.info()
```

```python
def get_dom (dt):
    return dt.day
```

```python
df["day"] = df["Date/Time"].map(get_dom)
```

```python
df
```

```python
def get_weekday (dt):
    return dt.weekday() #Weekday is a method
```

```python
df["Weekday"] = df["Date/Time"].map(get_weekday)
```

```python
df
```

```python
def get_hour (dt):
    return dt.hour #hour is an attribut
```

```python
df["hour"] = df["Date/Time"].map(get_hour)
```

```python
df
```

```python
df.describe()
```

```python
hist = df["day"].plot.hist(bins=30, rwidth=0.8, range=(0.5,30.5), title = "Frequency by DoM - Uber -April 2014")
plt.xlabel("Days of the month")
```

```python
def count_rows(rows):
    return len(rows)
```

```python
by_date = df.groupby("day").apply(count_rows)
by_date
```

```python
plt.title("Line plot - Uber - April 2014");
plt.xlabel("Days of the month")
plt.ylabel("Frequency")
plt.plot(by_date);
```
